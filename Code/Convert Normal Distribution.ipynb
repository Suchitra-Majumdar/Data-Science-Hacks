```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exponential Distribution
X = np.random.exponential(2,(10000,1))

# Log Normal Distrbution
Y = np.random.weibull(a=2,size=(10000,1))
```

### Plot Exponential and weibull distributions

```python
plt.figure(figsize=(10,4),dpi=120)

# Plotting exponential distrbution
plt.subplot(121)
plt.hist(X,bins=100)
plt.title("exponential distribtion")

#Plotting weibull ditribution
plt.subplot(122)
plt.hist(Y,bins=100)
plt.title("weibll distribution")
```

```python
from sklearn.preprocessing import QuantileTransformer
qt = QuantileTransformer(n_quantiles=len(X), output_distribution="normal")

## transforming above distributions to Normal Distribution
X = qt.fit_transform(X)
Y = qt.fit_transform(Y)
print("distributions transformed")
```

### Plot transformed distrbutions


```python
plt.figure(figsize=(10,4),dpi=120)

#Plotting transformed exponential
plt.subplot(121)
plt.hist(X,bins=100)
plt.title("transformed exponential")


# Plotting transformed weibull distributions
plt.subplot(122)
plt.hist(Y,bins=100)
plt.title("transformed weibull")
plt.show()
```
